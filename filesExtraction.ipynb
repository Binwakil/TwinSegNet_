```python
import os
import gzip
import shutil

# Function to extract .gz files
def extract_gz(file_path, output_path):
    with gzip.open(file_path, 'rb') as f_in:
        with open(output_path, 'wb') as f_out:
            shutil.copyfileobj(f_in, f_out)

# Path to dataset directory containing .gz files
data_dir = 'Data/2023GLI/BraTS2023-GLI--ValidationData'

# Iterate through dataset directory
for patient in os.listdir(data_dir):
    patient_path = os.path.join(data_dir, patient)
    print(patient_path)

    # Define the .gz files to extract
    for file_name in ['-flair.nii.gz',  '-t1.nii.gz', '-t1ce.nii.gz', '-t2.nii.gz', '-seg.nii.gz']:
        gz_file = os.path.join(patient_path, f'{patient}{file_name}')
        if os.path.isfile(gz_file):
            output_file = gz_file[:-3]  # Remove the .gz extension
            print(f"Extracting {gz_file}...")
            extract_gz(gz_file, output_file)
            print(f"Extracted to {output_file}")
            
            # Remove the .gz file after extraction
            os.remove(gz_file)
            print(f"Removed {gz_file}")
```

```python
import numpy as np
from skimage.transform import resize
import gc
import os

def preprocess_images(images, target_shape=(128, 128, 128)):
    preprocessed_images = []
    
    for img in images:
        # Resize each image to the target shape
        img_resized = resize(img, target_shape, mode='constant', anti_aliasing=True)
        
        # Normalize each image to [0, 1] range
        img_normalized = img_resized / np.max(img_resized)
        
        preprocessed_images.append(img_normalized)
    
    return np.array(preprocessed_images)

def process_and_save_images_in_batches(directory, output_dir, batch_size=10, target_shape=(128, 128, 128)):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)  # Create output directory if it doesn't exist
    
    patient_list = os.listdir(directory)
    
    for i in range(0, len(patient_list), batch_size):
        batch_patients = patient_list[i:i + batch_size]
        batch_images = []
        
        for patient in batch_patients:
            patient_path = os.path.join(directory, patient)
            flair_path = os.path.join(patient_path, f'{patient}_flair.nii')

            # Load the file if it exists
            if os.path.isfile(flair_path):
                flair_data = load_nii(flair_path)
                batch_images.append(flair_data)
                print(f"Processed patient: {patient}")
        
        if batch_images:
            # Preprocess the batch of images
            preprocessed_batch = preprocess_images(batch_images, target_shape)
            
            # Save the preprocessed batch as a .npy file
            output_file = os.path.join(output_dir, f'batch_{i//batch_size}.npy')
            np.save(output_file, preprocessed_batch)
            print(f"Saved batch {i//batch_size} to {output_file}")
            
            # Clear memory between batches
            gc.collect()

# Specify the directory and output path
input_directory = 'Data/BraTS2021'
output_directory = 'Data/Processed'

# Process and save the images in batches
process_and_save_images_in_batches(input_directory, output_directory, batch_size=5, target_shape=(128, 128, 128))
```
